```python
# Variables Acumuladoras
Acu =0

for i in range(0,11,2):
  #print(i)
  Acu=Acu+i
print(f"El Total de la suma de los numeros Pares ={Acu}")
```

```python
# Bucle Numeros hasta 100
Cuenta=0
for i in range(0,101,2):
  Cuenta=Cuenta+1
print(f"Hay= {Cuenta-1} Numeros excluyendo el cero")
print(f"Hay= {Cuenta} Numeros incluyendo el cero")
```

```python
# Tabla de Multiplicar
Ingrese_numero = int(input("Digite el número de la tabla de multiplicar: "))

for i in range(11):
  print(i, " x ", Ingrese_numero, " = ",  i * Ingrese_numero)
```
